```python
t=np.arange(0.,5,0.2)
plt.plot(t,t**2,'b--',label='^2')
plt.plot(t,t**2.2,'rs',label='^2.2')
plt.plot(t,t**2.5,'g^',label='^2.5')
plt.grid()
plt.legend()
plt.show()
```

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
```

```python
t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.2)
plt.figure(1)
plt.subplot(211)
plt.grid()
plt.plot(t1,f(t1),'b-')
plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.show()

```
